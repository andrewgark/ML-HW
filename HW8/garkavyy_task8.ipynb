```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * np.dot(F, np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(Sigma)

    # compute the value m: number of minumum features that retains the given variance varRetaine
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / \
                dTot * 100.0 for i in range(n)])
    for i, v in enumerate(var_i):
        if v > varRetained * 100:
            break
    m = np.argmax(var_i > varRetained * 100)
    print('%.2f variance retained in %d dimensions' % (varRetained, m))

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = U[:, :m]
    G = np.dot(U, np.diag(d))[:, :m]

    return G, U_reduced
```

### Кластеризация цифр с помощью dbscan
На данных из sklearn.datasets.load_digits примените алгоритмы кластеризации (знания о метках классов при кластеризации использовать нельзя):

 запускайте при различных параметрах eps и minsamples, для всех экспериментов можете выбрать одну метрику (вспомните семинар про  метрические алгоритмы);

 - визуалируйте изображения тех цифр, которые соответствуют core_points;
 - визуалируйте изображения тех цифр, которые соответствуют выбросам;
 - сделайте выводы и применимости алгоритмов.

```python
from sklearn import cluster, metrics
from sklearn.datasets import load_digits

data = load_digits()
digits = np.reshape(data.images, (data.images.shape[0], -1))
```

```python
def metrics_for_dbscan(x, *params):
    eps, min_samples = x
    digits = params[0]
    dbscan = cluster.DBSCAN(eps=eps, 
                            min_samples=min_samples, 
                            metric='euclidean', 
                            algorithm='auto', 
                            leaf_size=30, 
                            p=None, 
                            n_jobs=2)
    labels = dbscan.fit_predict(digits)
    return -(metrics.adjusted_mutual_info_score(labels, data.target) + 
             metrics.adjusted_rand_score(labels, data.target))
```

```python
from scipy.optimize import brute

def best_dbscan(eps_range, min_samples_range, data):
    res_brute = brute(metrics_for_dbscan,
                      (eps_range, 
                       min_samples_range),
                      args=(data,),
                      full_output=True,
                      finish=None)
    eps, min_samples = res_brute[0]
    return cluster.DBSCAN(eps=eps, 
                          min_samples=min_samples, 
                          metric='euclidean', 
                          algorithm='auto', 
                          leaf_size=30, 
                          p=None, 
                          n_jobs=2)
```

```python
dbscan =  best_dbscan(slice(15, 25, 0.5),
                      slice(2, 10, 1),
                      digits)
```

```python
labels = dbscan.fit_predict(digits)
print(metrics.adjusted_mutual_info_score(labels, data.target))
print(metrics.adjusted_rand_score(labels, data.target))
```

```python
def draw(images):
    plt.figure(figsize=(15, 10))
    for i, image in enumerate(images):
        plt.subplot(i // 10 + 1, 10, i + 1)
        plt.axis('off')
        plt.imshow(image)
    plt.show()
```

```python
print("core_points")
for i in range(2):
    draw(data.images[dbscan.core_sample_indices_[i * 10:(i + 1) * 10]])
```

```python
print("examples of -1")
for i in range(2):
    draw(data.images[labels == -1][i * 10:(i + 1) * 10])
```

```python
import cv2

def draw_picture(image, bgr=False):
    b,g,r = cv2.split(image) # bgr -> rgb
    new_image = cv2.merge([r,g,b])
    plt.figure(figsize=(7, 5))
    plt.axis('off')
    plt.imshow(new_image)
    plt.show()
    

img = cv2.imread('birds.jpg')
draw_picture(img)
```

Кажется, красный и синий цвета местами поменялись.

```python
from skimage import measure
from sklearn import cluster 

def draw_picture_with_model(model, img):
    linear_img = np.reshape(img, (-1, 3))
    labels = model.fit_predict(linear_img)
    values = model.cluster_centers_
    new_img = list(map(lambda x: values[labels[x[0]]], enumerate(linear_img)))
    new_img = np.reshape(new_img, img.shape)
    new_img = new_img.astype(img.dtype)
    print('%d clusters' % len(np.unique(labels)))
    print('SSIM = %f' % measure.compare_ssim(new_img, img, multichannel=True))
    draw_picture(new_img)
```

```python
for n_clusters in [2, 3, 10, 20]:
    draw_picture_with_model(cluster.KMeans(n_clusters=n_clusters,
                                           max_iter=10,
                                           n_init=10,
                                           n_jobs=2),
                            img)
```

```python
for quantile in [0.2, 0.1, 0.05, 0.01]:
    draw_picture_with_model(cluster.MeanShift(bandwidth=cluster.estimate_bandwidth(np.reshape(img, (-1, 3)),
                                                                                   quantile=quantile, 
                                                                                   n_samples=500),
                                              bin_seeding=True,
                                              n_jobs=2),
                            img)
```

Давайте посмотрим на такие же числа кластеров, но только в k-means:

```python
for n_clusters in [6, 11, 180]:
    draw_picture_with_model(cluster.KMeans(n_clusters=n_clusters,
                                           max_iter=10,
                                           n_init=10,
                                           n_jobs=2),
                            img)
```

Мы явно видим, что K-Means работает лучше, чем MeanShift (и по SSIM, и по качеству картинок).
